```python
import mxnet as mx
from longling.ML.MxnetHelper.toolkit.lr_scheduler import plot_schedule
schedule = mx.lr_scheduler.FactorScheduler(base_lr=0.01, step=10, factor=0.5, warmup_steps=20)
lrs = plot_schedule(schedule, 40)
lrs[-1]
```

```python
from longling.ML.MxnetHelper import get_lr_scheduler

scheduler = get_lr_scheduler("factor", base_lr=0.01, step=10, factor=0.5, warmup_steps=5)
lrs = plot_schedule(scheduler, 40)
lrs[-1]
```

```python
scheduler = get_lr_scheduler("factor", base_lr=0.01, step=2, max_update=20, discount=0.1, warmup_steps=10)
lrs = plot_schedule(scheduler, 20)
lrs[-1]
```

```python
scheduler = get_lr_scheduler("linear", base_lr=0.01, max_update=20, discount=0.1, warmup_steps=10)
lrs = plot_schedule(scheduler, 20)
lrs[-1]
```

```python
scheduler = get_lr_scheduler("linear", base_lr=0.01, max_update=20, discount=0.1, warmup_steps=10, step=2)
lrs = plot_schedule(scheduler, 20)
lrs[-1]
```

```python
scheduler = get_lr_scheduler("linear", lr=0.01, discount=0.1, batches_per_epoch=10, update_epoch=2, warmup_epoch=1)
lrs = plot_schedule(scheduler, 30)
lrs[-1]
```

```python
scheduler = get_lr_scheduler("linear", lr=0.01, discount=0.1, batches_per_epoch=10, update_epoch=2, warmup_epoch=1, epoch_update_freq=0)
lrs = plot_schedule(scheduler, 40)
lrs[-1]
```
